```r

library(tidyverse)
library(repr)
library(dbplyr)
```

```r
main_data_column_unnamed<- read_delim("cleve.mod", col_names = FALSE,  delim = ' ', skip = 20)
main_data_column_unnamed
```

```r
main_data_column_2 <- read_table("cleve.mod", col_names = c('Age', 'Sex', 'Chest Pain Type', 'Resting Blood Pressure',
                                                            'Cholesterol', 'Fasting Blood Sugar <120',
                                                        'Resting ECG Reading', 'Max Heart Rate',
                                                       'Exercise Induced Angina (TRUE or FALSE)',
                                                            'Old Peak', 'Slope', 'Number Of Vessels Coloured', 'thal'), skip = 20)

main_data_column_2
```

```r
quick_summary <- head(main_data_column_2)
quick_summary
```

Checking **Average Age** of both Male and Female Patients

```r
average_data_female_age <- main_data_column_2 |>
group_by(Sex = 'fem') |>
summarize(Age = mean(Age, na.rm = TRUE))


average_data_female_age
```

```r
average_data_male_age <- main_data_column_2 |>
group_by(Sex = 'male') |>
summarize(Age = mean(Age, na.rm = TRUE))

average_data_male_age
```
