```r
library(tidyverse)
library(janitor)
```

```r
main_data_column_2 <- read_table("cleve.mod", col_names = c('Age', 'Sex', 'Chest Pain Type', 'Resting Blood Pressure',
                                                            'Cholesterol', 'Fasting Blood Sugar <120',
                                                        'Resting ECG Reading', 'Max Heart Rate',
                                                       'Exercise Induced Angina (TRUE or FALSE)',
                                                            'Old Peak', 'Slope', 'Number Of Vessels Coloured', 'thal','Health'), skip = 20)
main_data_column_2

main_data_column_2<-clean_names(main_data_column_2)
main_data_column_2
colnames(main_data_column_2)
```

```r
heart_data_plot<-main_data_column_2|>
    ggplot(aes(x=age,y=cholesterol,color=health))+
    geom_point()+
    labs(x="Age",y="Cholesterol",color="Health")
heart_data_plot
```
